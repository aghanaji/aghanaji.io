```python
from module.hashem import *
from module.stg_def import *
from module.stg import *

mt5.initialize()

total_balance = 5000
daily_drow_down = 0.02
full_drow_down = 0.10
daily_target = 0.04
today_balance = balance()

symbols = ['XAUUSD.']

while True :


    if check_time(11,13):
        ORG_risk = 0.1
    else:
        ORG_risk = 0.05


    if draw_down_checker(today_balance , pnl_today(3) , daily_drow_down) or total_draw_down(total_balance , full_drow_down):
        close_all_positions()
        mt5.shutdown()
        raise Exception('max dd hit')


    if pnl_today(3) > (today_balance * daily_target):
        close_all_positions()
        mt5.shutdown()
        raise Exception('max target profit today')

        

    for i in symbols:

        symbol = f'{i}'

        from module.stg_var import *
        
        if count_sl_in_hours(1) > 1 and count_sl() > count_tp():
            verify3m = 1
            verify5m = 1


        if (count_sl() >= 4 and count_sl() > count_tp() ) or (count_sl() > (count_tp() * 2) and count_sl() > 2 ) :
            verify3m = 1 
            verify5m = 1
          


        if is_news():
            verify3m = 1
            verify5m = 1
            close_all_positions()
            


        variables = ['halfXAU5b' , 'halfXAU5s' , 'halfXAU3b' , 'halfXAU3s']


        for variable_name in variables:

            if count_sl_in_hours_with_comment(variable_name, hours=1) >= 2:
                globals()[variable_name] = 1

            if count_sl_with_comment(variable_name) > 3 :
                globals()[variable_name] = 1



            
        def position_modify(position):
            
            globals()[position._asdict()['comment']] = 1
            position_symbol = position._asdict()['symbol']
            price_ask = mt5.symbol_info_tick(position_symbol).ask
            price_bid = mt5.symbol_info_tick(position_symbol).bid

            if position._asdict()['type'] == 0 :
                
                if half_trend(position_symbol , '3m' )[-2] == 'short':
                    close_all_with_comment(position._asdict()['comment'])

                    orders_total = mt5.orders_get()
                    for order in orders_total:
                        if order._asdict()['comment'] == position._asdict()['comment']:
                            remove_order(position_symbol , order._asdict()['ticket'])
                                        
                    remove_order

                if position._asdict()['price_open'] - position._asdict()['sl'] <= price_ask -  position._asdict()['price_open'] and position._asdict()['sl'] < position._asdict()['price_open']:
                    modify_position(position._asdict()['ticket'] , ((price_ask -  position._asdict()['sl']) / 3 ))
                
                if position._asdict()['price_open'] <= position._asdict()['sl'] :
                    modify_profit(position , price_ask)


            if position._asdict()['type'] == 1 :
                
                if half_trend(position_symbol , '3m' )[-2] == 'long':
                    close_all_with_comment(position._asdict()['comment'])
                    orders_total = mt5.orders_get()
                    for order in orders_total:
                        if order._asdict()['comment'] == position._asdict()['comment']:
                            remove_order(position_symbol , order._asdict()['ticket'])


                if position._asdict()['sl'] - position._asdict()['price_open'] <=  position._asdict()['price_open'] - price_bid and position._asdict()['sl'] > position._asdict()['price_open']:
                    modify_position(position._asdict()['ticket'] , ((position._asdict()['sl'] - price_bid) / 3 ))
                
                if position._asdict()['price_open'] >= position._asdict()['sl'] :
                    modify_profit(position , price_ask)


        positions = mt5.positions_get()

        for position in positions :
            position_modify(position)
            


        if check_time(13 , 19):
            verify5m = 1

        #5m -----------------------------------------------------------------------
        if verify5m == 0 :
            parabolic_STG(symbol , '5m' , 'halfXAU5b' , 'halfXAU5s' ,ORG_risk , halfXAU5b , halfXAU5s)


        if verify3m == 0 :
            parabolic_STG(symbol , '3m' , 'halfXAU3b' , 'halfXAU3s' ,ORG_risk , halfXAU3b , halfXAU3s)


```
